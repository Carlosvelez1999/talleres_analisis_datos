```python
# Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
# Cargar el archivo de Excel
df = pd.read_excel('Datos_1.xlsx', sheet_name = 'frame_section_assigments')
```

```python
# Mostrar las primeras filas del dataframe
print(df.head())
```

```python
# Identificar las variables del tipo categoricas
object_columns = df.select_dtypes(include=['object']).columns
print("Variables categóricas:", object_columns)
```

```python
# Análisis univariado
for column in object_columns:
    # Distribución de frecuencias
    print(f"Distribución de {column}:")
    print(df[column].value_counts())
    print("\n")
    
    # Visualización
    plt.figure(figsize=(10, 5))
    sns.countplot(x=column, data=df, order=df[column].value_counts().index)
    plt.title(f"Distribución de {column}")
    plt.xticks(rotation=45)
    plt.show()
    
    # Cantidad de categorías únicas
    unique_values = df[column].nunique()
    print(f"La variable {column} tiene {unique_values} categorías únicas.\n")
```

```python
# Verificación de valores faltantes
print("Valores faltantes por columna:")
print(df[object_columns].isnull().sum())
```
